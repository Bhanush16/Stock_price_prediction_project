#Data preprocessing

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df=df.sort_values('Date')
```

```python
df.set_index('Date',inplace=True)
```

```python
df.dropna()
```

```python
df.fillna(method='ffill')
```

```python
df['MA10'] = df['Close'].rolling(window=10).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()
```

```python
df['Return'] = df['Close'].pct_change()
```

```python
df['Volatility'] = df['Return'].rolling(window=10).std()
```

```python
df['Lag1'] = df['Close'].shift(1)
df['Lag2'] = df['Close'].shift(2)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Close']] = scaler.fit_transform(df[['Close']])
```

```python
df
```

```python
train_size = int(len(df) * 0.8)
train = df[:train_size]
test = df[train_size:]
```

```python
train = train.dropna()
test = test.dropna()
X_train = train[['MA10', 'MA50', 'Volatility', 'Lag1', 'Lag2']]
y_train = train['Close']
X_test = test[['MA10', 'MA50', 'Volatility', 'Lag1', 'Lag2']]
y_test = test['Close']
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
X_train = train[['MA10', 'MA50', 'Volatility', 'Lag1', 'Lag2']]
y_train = train['Close']
X_test = test[['MA10', 'MA50', 'Volatility', 'Lag1', 'Lag2']]
y_test = test['Close']
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print("MSE:", mse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test, label="Actual Prices", color="black")
plt.plot(y_test.index, y_pred, label="Predicted Prices", color="green", linestyle="--")
plt.title("Actual vs Predicted Stock Prices (Linear Regression)")
plt.xlabel("Date")
plt.ylabel("Price (scaled)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_test_original = scaler.inverse_transform(y_test.values.reshape(-1, 1))
y_pred_original = scaler.inverse_transform(y_pred.reshape(-1, 1))
plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test_original, label="Actual Prices", color="black")
plt.plot(y_test.index, y_pred_original, label="Predicted Prices", color="green", linestyle="--")
plt.title("Actual vs Predicted Stock Prices (Linear Regression)")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()

```

#Comparing models

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
y_test_original = scaler.inverse_transform(y_test.values.reshape(-1, 1))
y_pred_rf_original = scaler.inverse_transform(y_pred_rf.reshape(-1, 1))


mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Results:")
print("MSE:", mse_rf)
print("RMSE:", rmse_rf)
print("R² Score:", r2_rf)

plt.figure(figsize=(14,7))
plt.plot(y_test.index, y_test_original, label="Actual Prices", color="black")
plt.plot(y_test.index, y_pred_rf_original, label="Predicted Prices (RF)", color="blue", linestyle="--")
plt.title("Actual vs Predicted Stock Prices (Random Forest)")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVR

# Train Support Vector Regressor
svm = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.01)
svm.fit(X_train, y_train)

# Predictions
y_pred_svm = svm.predict(X_test)

# Evaluation
mse_svm = mean_squared_error(y_test, y_pred_svm)
rmse_svm = np.sqrt(mse_svm)
r2_svm = r2_score(y_test, y_pred_svm)

print("SVM Results:")
print("MSE:", mse_svm)
print("RMSE:", rmse_svm)
print("R² Score:", r2_svm)

# Add to comparison table
results = {
    "Model": ["Linear Regression", "Random Forest", "SVM"],
    "MSE": [mse, mse_rf, mse_svm],
    "RMSE": [rmse, rmse_rf, rmse_svm],
    "R² Score": [r2, r2_rf, r2_svm]
}
results_df = pd.DataFrame(results)
results_df
```

```python
from sklearn.tree import DecisionTreeRegressor

# Train Decision Tree
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)

# Predictions
y_pred_dt = dt.predict(X_test)

# Evaluation
mse_dt = mean_squared_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Results:")
print("MSE:", mse_dt)
print("RMSE:", rmse_dt)
print("R² Score:", r2_dt)
```

```python
from xgboost import XGBRegressor

# Train XGBoost
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
xgb.fit(X_train, y_train)

# Predictions
y_pred_xgb = xgb.predict(X_test)

# Evaluation
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print("XGBoost Results:")
print("MSE:", mse_xgb)
print("RMSE:", rmse_xgb)
print("R² Score:", r2_xgb)
```
